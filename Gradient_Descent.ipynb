```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(grad, w_init, learning_rate=0.1, n_steps=20):
  w = np.array(w_init)
  path = [w.copy()]
  for step in range(n_steps):
    w = w - learning_rate * grad(w)
    path.append(w.copy())
  return np.array(path)
```

# Define a Convex Cost Function

```python
def J_convex(w):
  return w[0]**2 + 2 * w[1]**2

def grad_J_convex(w):
  return np.array([2 * w[0], 4 * w[1]])
```

# Visualize the Gradient Descent Path

```python
w_init = [4, 4]
path_convex = gradient_descent(grad_J_convex, w_init, learning_rate=0.05, n_steps=1000)

w1_range = np.linspace(-5, 5, 400)
w2_range = np.linspace(-5, 5, 400)
W1, W2 = np.meshgrid(w1_range, w2_range)
J = J_convex([W1, W2])

plt.figure(figsize=(8, 6))
contour_level = np.logspace(-0.5, 3, 35)
plt.contour(W1, W2, J, levels=contour_level, cmap='jet')
path_x, path_y = path_convex[:, 0], path_convex[:, 1]
plt.plot(path_x, path_y, marker='o', color='red', label='Gradient Descent Path')
plt.title('Gradient Descent on a Convex Cost Function')
plt.xlabel('$w_1$')
plt.ylabel('$w_2$')
plt.legend()
plt.show()
```
